# Importing data and libraries

```python
import numpy as np
import pandas as pd 

import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
df = pd.read_excel('/Users/selim/Desktop/MA1/ML/Local tests/Sentiment Analysis/SA_Vader.xlsx')
```

Remark: Change file paths according to your own directories in order to assure smooth code execution 

# Sentiment analysis

```python
# We add a new columns that holds the vader scores for each articles
df['scores'] = df['text'].apply(lambda review: sid.polarity_scores(str(review)))
```

```python
# We add a new column that holds the vader compound score
df['compound'] = df['scores'].apply(lambda score_dict: score_dict['compound'])
```

```python
# We add a new column that holds the overall label of the article
df['label'] = df['compound'].apply(lambda c: 'positive' if c > 0 else 'negative')
```

# Save output to xlsx

```python
df.to_excel('/Users/selim/Desktop/MA1/ML/Local tests/Sentiment Analysis/SA_Vader.xlsx', index=False)
```
