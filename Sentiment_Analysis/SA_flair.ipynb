# Sentiment analysis

```python
# Creates an english sentiment model
sentiment_model = flair.models.TextClassifier.load('en-sentiment')
```

```python
# Function that does sentiment analysis on text and extracts the polarity and confidence
sentiment = []
confidence = []

def sent_analysis(text):
  sample = flair.data.Sentence(str(text))
  sentiment_model.predict(sample)
  sentiment.append(sample.labels[0].value)
  confidence.append(sample.labels[0].score)
  return text
```

```python
# Applying sentiment analysis on the articles
df['text'] = df['text'].apply(sent_analysis)
```

```python
# Adding new columns in the dataframe to hold the sentiment and confidence 
df['sentiment'] = sentiment
df['confidence'] = confidence 
```
