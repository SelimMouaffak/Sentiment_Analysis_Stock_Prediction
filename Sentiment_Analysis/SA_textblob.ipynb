# Sentiment analysis

```python
# Function that extracts the polarity of a text
def get_polarity(text):
  return TextBlob(str(text)).sentiment.polarity
```

```python
# Function that extracts the subjectivity of a text
def get_subjectivity(text):
  return TextBlob(str(text)).sentiment.subjectivity
```

```python
# Function that crafts an overall label for the text
def get_label(score):
  if score < 0:
    return 'Negative'
  elif score == 0:
    return 'Neutral'
  else:
    return 'Positive'
```

```python
# We add new columns that hold the polarity and subjectivity of each article
df['polarity'] = df['text'].apply(get_polarity)
df['subjectivity'] = df['text'].apply(get_subjectivity)
```

```python
# We add a new column that hold the overall label for each article
df['label'] = df['polarity'].apply(get_label)
```
