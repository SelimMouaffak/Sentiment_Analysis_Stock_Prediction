# Importing data and libraries

```python
import numpy as np
import pandas as pd 

import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
df = pd.read_excel('/content/drive/MyDrive/Sentiment_Analysis_Stock_Prediction/DATA/clean_dataset.xlsx')
```

```python
df_orig = df
```

```python
df = df_orig
```

# Sentiment analysis

```python
df['scores'] = df['text'].apply(lambda review: sid.polarity_scores(str(review)))
```

```python
df['compound'] = df['scores'].apply(lambda score_dict: score_dict['compound'])
```

```python
df['label'] = df['compound'].apply(lambda c: 'positive' if c > 0 else 'negative')
```

# Save output to xlsx

```python
df.to_excel('/content/drive/MyDrive/Sentiment_Analysis_Stock_Prediction/DATA/SA_Vader.xlsx', index=False)
```
